```python
import camelot
import pandas as pd
```

```python
def data_process(df:pd.DataFrame):
#     必要なカラムと欠損の補完,削除(いらないかも)
    return df

def fillempty(df:pd.DataFrame)->pd.DataFrame:#利用日の空白を補完
    day = df.iloc[0,0]
    for i in range(len(df)):
        if(df.iloc[i,0]!=''):
            day = df.iloc[i,0]
        df.iloc[i,0] = day
    return df

def split_day_time(df : pd.DataFrame)->pd.DataFrame:#利用日と時間が結合されているセルを分離
#     利用日と時間が結合されているものとそうでないものに分ける
    df_err = df.query('time==""')
    df = df.query('time!=""')
#     splitで"day"カラムを分割し再代入(順序逆にしたらバグる)
    df_err.time = df_err.day.apply(lambda x :x.split(' ')[1])
    df_err.day = df_err.day.apply(lambda x :x.split(' ')[0])
#     再度結合し、"day" と"time"でソートする
    df = pd.concat([df,df_err],axis=0)
    df = df.sort_values(['day','time'])
    return df
```
